```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_auto = pd.read_csv('Week 2')
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df_auto.columns = headers
```

```python
df_auto.head()
```

```python
df_auto.replace('?', np.nan, inplace = True)
```

```python
df_auto.head()
```

```python
avg_normloss = df_auto['normalized-losses'].astype('float').mean(axis=0)
avg_normloss
```

```python
df_auto['normalized-losses'].replace(np.nan, avg_normloss,inplace = True)
df_auto.head()
```

```python
avg_stroke = df_auto['stroke'].astype('float').mean(axis=0)
avg_stroke
```

```python
def mean_replace():
    col_change = ['stroke','bore','horsepower','peak-rpm']
    for i in col_change:
        avg = df_auto[i].astype('float').mean(axis=0)
        df_auto[i].replace(np.nan, avg,inplace = True)
    return df_auto
```

```python
avg_stroke = df_auto['stroke'].astype('float').mean(axis=0)
avg_stroke
```

```python
df_auto['city-L/100km'] = 235/df_auto["city-mpg"]
df_auto.head()
```

```python


# transform mpg to L/100km by mathematical operation (235 divided by mpg)
#df_auto["highway-mpg"] = 235/df_auto["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
#df_auto.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df_auto["highway-L/100km"] = 235/df_auto["highway-L/100km"]
# check your transformed data 
df_auto.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df_auto['length'] = df_auto['length']/df_auto['length'].max()
df_auto['width'] = df_auto['width']/df_auto['width'].max()
```

```python
df_auto['height'] = df_auto['height']/df_auto['height'].max() 

# show the scaled columns
df_auto[["length","width","height"]].head()
```

```python
df_auto["horsepower"]=df_auto["horsepower"].astype(int, copy=True)
```

```python
plt.hist(df_auto["horsepower"])

# set x/y labels and plot title
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bin")
```

```python
bins = np.linspace(min(df_auto["horsepower"]), max(df_auto["horsepower"]), 4)
bins
```

```python
df_auto['horsepower-binned'] = pd.cut(df_auto['horsepower'], bins, labels=group_names, include_lowest=True )
df_auto[['horsepower','horsepower-binned']].head(20)
```

```python
dummy_variable_1 = pd.get_dummies(df_auto["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# merge data frame "df" and "dummy_variable_1" 
df_auto = pd.concat([df_auto, dummy_variable_1], axis=1)

# drop original column "fuel-type" from "df"
df_auto.drop("fuel-type", axis = 1, inplace=True)
```

```python
df_auto.head()
```

```python
dummy_variable_2 = pd.get_dummies(df_auto['aspiration'])
dummy_variable_2
```

```python
dummy_variable_2.rename(columns={'std':'aspiration-std', 'turbo': 'aspiration-turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
df_auto = pd.concat([df_auto, dummy_variable_2], axis=1)
df_auto.drop('aspiration', axis = 1,inplace = True)
df_auto.head()
```
